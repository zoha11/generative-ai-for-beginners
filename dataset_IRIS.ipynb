```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris.data.shape
iris.feature_names
```

```python
iris.target_names
```

```python
iris.data[0:5]
```

```python
iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df.head()
```

```python
pd.plotting.scatter_matrix(iris_df,c=iris.target,figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8)
```

```python
x = iris.data
y= iris.target
```

```python
plt.scatter(x[:,0],x[:,1], c=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
```

```python
x= iris.data
y= iris.target
```

```python
knn.fit(x,y)
```

```python
y_fg = knn.predict(np.array([[3,4,6.5,0.2]]))
y_fg
```
